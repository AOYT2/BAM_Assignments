```python
import gurobipy as gp
```

```python
model=gp.Model()

tNO=model.addVar(lb=0,ub=10000,name="t2no")
tChar=model.addVar(lb=0,ub=10000,name="t2char")
tSea=model.addVar(lb=0,ub=10000,name="t2sea")
model.addConstr(tNO+tChar+tSea<=10000,name="Texas Suppy")

cNO=model.addVar(lb=0,ub=50000,name="c2no")
cChar=model.addVar(lb=0,ub=50000,name="c2char")
cSea=model.addVar(lb=0,ub=50000,name="c2sea")
model.addConstr(cNO+cChar+cSea<=50000,name="California Suppy")

dPit=model.addVar(20000,name="demand pit")
dAt=model.addVar(25000,name="demand at")

costTex=(11*tNO+7*tChar+2*tSea)
costCal=(7*cNO+4*cChar+8*cSea)

supplyNO=(tNO+cNO)
supplyChar=(tChar+cChar)
supplySea=(tSea+cSea)

noPit=model.addVar(lb=0,ub=60000,name="no2pit")
noAt=model.addVar(lb=0,ub=60000,name="no2at")
model.addConstr(noPit+noAt<=supplyNO,name="New Orleeens Supply")

charPit=model.addVar(lb=0,ub=60000,name="char2pit")
charAt=model.addVar(lb=0,ub=60000,name="char2at")
model.addConstr(charPit+charAt<=supplyChar,name="Charlton Supply")

seaPit=model.addVar(lb=0,ub=60000,name="sea2pit")
seaAt=model.addVar(lb=0,ub=60000,name="sea2at")
model.addConstr(seaPit+seaAt<=supplySea,name="Seatle Supply")

costNO=(11*noPit+7*noAt)
costChar=(7*charPit+4*charAt)
costSea=(5*seaPit+3*seaAt)

model.addConstr((noPit+charPit+seaPit)==dPit,name="Pitssburgh Delivery")
model.addConstr((noAt+charAt+seaAt)==dAt,name="Atlanta Delivery")

model.setObjective(costTex+costCal+costNO+costChar+costSea,gp.GRB.MINIMIZE)
model.optimize()
if not model.status == gp.GRB.OPTIMAL:
    print("something went wrong")
print("Total cost is",model.objval)
model.printAttr("X")
```

```python
model = gp.Model()

#purchase
PA = model.addVar(lb = 0, ub = 100, name = 'PA')
PB = model.addVar(lb = 0, ub = 100, name = 'PB')

#A2B
EA2B_1 = model.addVar(lb = 0, name = 'EA2B_1')
EA2B_2 = model.addVar(lb = 0, name = 'EA2B_2')
EA2B_3 = model.addVar(lb = 0, name = 'EA2B_3')
EA2B_4 = model.addVar(lb = 0, name = 'EA2B_4')
EA2B_5 = model.addVar(lb = 0, name = 'EA2B_5')
EA2B_6 = model.addVar(lb = 0, name = 'EA2B_6')
EA2B_7 = model.addVar(lb = 0, name = 'EA2B_6')
CA2B_1 = model.addVar(lb = 0, name = 'CA2B_1')
CA2B_2 = model.addVar(lb = 0, name = 'CA2B_2')
CA2B_3 = model.addVar(lb = 0, name = 'CA2B_3')
CA2B_4 = model.addVar(lb = 0, name = 'CA2B_4')
CA2B_5 = model.addVar(lb = 0, name = 'CA2B_5')
CA2B_6 = model.addVar(lb = 0, name = 'CA2B_6')
CA2B_7 = model.addVar(lb = 0, name = 'CA2B_7')

#B2A
EB2A_1 = model.addVar(lb = 0, name = 'EB2A_1')
EB2A_2 = model.addVar(lb = 0, name = 'EB2A_2')
EB2A_3 = model.addVar(lb = 0, name = 'EB2A_3')
EB2A_4 = model.addVar(lb = 0, name = 'EB2A_4')
EB2A_5 = model.addVar(lb = 0, name = 'EB2A_5')
EB2A_6 = model.addVar(lb = 0, name = 'EB2A_6')
EB2A_7 = model.addVar(lb = 0, name = 'EB2A_7')
CB2A_1 = model.addVar(lb = 0, name = 'CB2A_1')
CB2A_2 = model.addVar(lb = 0, name = 'CB2A_2')
CB2A_3 = model.addVar(lb = 0, name = 'CB2A_3')
CB2A_4 = model.addVar(lb = 0, name = 'CB2A_4')
CB2A_5 = model.addVar(lb = 0, name = 'CB2A_5')
CB2A_6 = model.addVar(lb = 0, name = 'CB2A_6')
CB2A_7 = model.addVar(lb = 0, name = 'CB2A_7')

#inventory
IA_1 = PA
IB_1 = PB
model.addConstr(IA_1 >= 45, name='IA_1_C')
model.addConstr(IB_1 >= 8, name='IAB_1_C')

#2
IA_2 = IA_1 + EB2A_1 - EA2B_1 - CA2B_1
IB_2 = IB_1 + EA2B_1 - EB2A_1 - CB2A_1
model.addConstr(IA_2 >= 20, name='IA_2_C')
model.addConstr(IB_2 >= 12, name='IB_2_C')

#3
IA_3 = IA_2 + EB2A_2 + CB2A_1 - EA2B_2 - CA2B_2
IB_3 = IB_2 + EA2B_2 + CA2B_1 - EB2A_2 - CB2A_2
model.addConstr(IA_3 >= 20, name='IA_3_C')
model.addConstr(IB_3 >= 23, name='IB_3_C')

#4
IA_4 = IA_3 + EB2A_3 + CB2A_2 - EA2B_3 - CA2B_3
IB_4 = IB_3 + EA2B_3 + CA2B_2 - EB2A_3 - CB2A_3
model.addConstr(IA_4 >= 25, name='IA_4_C')
model.addConstr(IB_4 >= 30, name='IB_4_C')

#5
IA_5 = IA_4 + EB2A_4 + CB2A_3 - EA2B_4 - CA2B_4
IB_5 = IB_4 + EA2B_4 + CA2B_3 - EB2A_4 - CB2A_4
model.addConstr(IA_5 >= 15, name='IA_5_C')
model.addConstr(IB_5 >= 12, name='IB_5_C')

#6
IA_6 = IA_5 + EB2A_5 + CB2A_4 - EA2B_5 - CA2B_5
IB_6 = IB_5 + EA2B_5 + CA2B_4 - EB2A_5 - CB2A_5
model.addConstr(IA_6 >= 28, name='IA_6_C')
model.addConstr(IB_6 >= 10, name='IB_6_C')

#7
IA_7 = IA_6 + EB2A_6 + CB2A_5 - EA2B_6
IB_7 = IB_6 + EA2B_6 + CA2B_5 - EB2A_6
model.addConstr(IA_7 >= 15, name='IA_7_C')
model.addConstr(IB_7 >= 33, name='IB_7_C')

#cost
purchase = 200*(PA+PB)
expensive = 20*(EA2B_1+EA2B_2+EA2B_3+EA2B_4+EA2B_5+EA2B_6+EB2A_1+EB2A_2+EB2A_3+EB2A_4+EB2A_5+EB2A_6)
cheap = 5*(CA2B_1+CA2B_2+CA2B_3+CA2B_4+CA2B_5+CA2B_6+CB2A_1+CB2A_2+CB2A_3+CB2A_4+CB2A_5+CB2A_6)

#objective
model.setObjective(purchase + expensive + cheap ,gp.GRB.MINIMIZE)

#finish
model.optimize()
if not model.status == gp.GRB.OPTIMAL:
    print("something went wrong")
print("Optimal cost is", model.objval)
model.printAttr("X")

```

```python
model = gp.Model()

#purchase
PA = model.addVar(lb = 0, ub = 100, name = 'PA',obj=200)
PB = model.addVar(lb = 0, ub = 100, name = 'PB',obj=200)

#basics
periods=7
demandA=[45,20,20,25,15,28,15]
demandB=[8,12,23,30,12,10,33]
expensive=20
cheap=5

#costs
EA2B=model.addVars(periods,name="ExpensiveA2B",obj=expensive)
CA2B=model.addVars(periods,name="cheapA2B",obj=cheap)

EB2A=model.addVars(periods,name="ExpensiveB2A",obj=expensive)
CB2A=model.addVars(periods,name="cheapB2A",obj=cheap)

#inventory
IA=model.addVars(periods,name="IA")
IB=model.addVars(periods,name="IB")

#constraints
model.addConstrs(IA[x]>=demandA[x] for x in range(periods))
model.addConstrs(IB[x]>=demandB[x] for x in range(periods))
model.addConstr(IA[0]==PA)
model.addConstr(IB[0]==PB)

model.addConstrs(IA[x]==IA[x-1]+EB2A[x]-EA2B[x]-CA2B[x] for x in range(1,2))
model.addConstrs(IB[x]==IB[x-1]+EA2B[x]-EB2A[x]-CB2A[x] for x in range(1,2))

model.addConstrs(IA[x]==IA[x-1]+EB2A[x]-EA2B[x]+CB2A[x-1]-CA2B[x] for x in range(2,periods))
model.addConstrs(IB[x]==IB[x-1]+EA2B[x]-EB2A[x]+CA2B[x-1]-CB2A[x] for x in range(2,periods))

#finish
model.optimize()
if not model.status == gp.GRB.OPTIMAL:
    print("something went wrong")
print("Optimal cost is", model.objval)
model.printAttr("X")
```
